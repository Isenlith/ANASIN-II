```
import numpy as np
from matplotlib import pyplot as plt
```

##Exemplo 01

```
t = np.linspace(-6, 6, 13/0.1)
N = 100
fs = 1


plt.subplot(2,1,1)
for n in range(0, N):
  s = (8/np.pi**2) * np.cos((2*n - 1)*np.pi*t/2)/(2*n-1)**2
  fs -= s
  plt.plot(t,s)
  
plt.show()

plt.subplot(2,1,2)
plt.plot(fs)
```

##Exemplo 02

```
t = np.linspace(-6, 6, 13/0.1)
N = 1000
fs = 1/2


plt.subplot(2,1,1)
for n in range(0, N):
  s = (2*(-1)**(n+1))/(np.pi*(2*n-1)) * np.cos(((2*n-1)*np.pi*t)/2)
  fs += s
  plt.plot(t,s)
  
plt.show()

plt.subplot(2,1,2)
plt.plot(t,fs)
```

```
fs = 200
tf = 1
t = np.arange(0, tf - 1/fs, 1/fs)
N = len(t)

freq = np.fft.fftfreq(N)
mascara = freq > 0


f = 3*np.cos(2*2*np.pi*t+np.pi/3) + 2*np.sin(2*np.pi*t) + 3*np.cos(2*np.pi*10*t)
plt.figure(1)
plt.plot(t, f)

F = np.fft.fft(f)
F_mag = np.abs(F)
F_phase = np.angle(F)
m = np.arange(0, N)
res = m*fs/len(F)

plt.figure(2)
plt.stem(res[mascara], F_mag[mascara]/N)

plt.figure(3)
plt.stem(res, F_phase)

plt.show()
```
