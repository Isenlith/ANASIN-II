```
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal
```

```
t = np.linspace(0,1,1000)

f = 1
x = np.sin(f*2*np.pi*t) + np.random.randn(t.shape[0])*0.2

plt.title("Sinal Original")
plt.plot(x);
```

##Filtro Passa Baixa

```
y = []
k = 100

for n in range(len(x)):
  s = 0
  for i in range(k):
    if n - i < 0:
      s += 0
    else:
      s += x[n - i]
  y.append(s/k)  
    
plt.plot(x);
plt.plot(y);
```

##Filtro passa alta

```
y = []
k = 100

for n in range(len(x)):
  s = 0
  for i in range(k):
    if n - i < 0:
      s -= 0
    elif i == 0:
      s += x[n - i]
    else:
      s -= x[n - i]
  y.append(s/k)

plt.plot(x);
plt.plot(y);
```
