```
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal
```

```
plt.rcParams["figure.figsize"] = [15, 5]

t = np.linspace(0,2,1000)

f = 1
sinal = np.sin(f*2*np.pi*t)
x = sinal + np.random.randn(t.shape[0])*0.2

plt.subplot(1,2,1)
plt.title("Sinal Original")
plt.plot(sinal);
plt.subplot(1,2,2)
plt.title("Sinal ruidoso")
plt.plot(x);
```

##Filtro Passa Baixa

#$y[n] = \frac{x[n] + x[n - 1] + ... + x[n - k]}{k}$



```
plt.rcParams["figure.figsize"] = [7, 5]

y = []
k = 100

for n in range(len(x)):
  s = 0
  for i in range(k):
    if n - i < 0:
      s += 0
    else:
      s += x[n - i]
  y.append(s/k)  
    
plt.plot(x);
plt.plot(y);
```

```
plt.rcParams["figure.figsize"] = [15, 5]

t = np.linspace(0,2,1000)

f = 10
sinal = np.sin(f*2*np.pi*t)
x = sinal + np.random.randn(t.shape[0])*0.7

plt.subplot(1,2,1)
plt.title("Sinal Original")
plt.plot(sinal);
plt.subplot(1,2,2)
plt.title("Sinal ruidoso")
plt.plot(x);
```

##Filtro passa alta

#$y[n] = \frac{x[n] - x[n - 1] - ... - x[n - k]}{k}$

```
plt.rcParams["figure.figsize"] = [7, 5]

y = []
k = 15

for n in range(len(x)):
  s = 0
  for i in range(k):
    if n - i < 0:
      s -= 0
    elif i == 0:
      s += x[n - i]
    else:
      s -= x[n - i]
  y.append(s/k)

plt.plot(x);
plt.plot(y);
```
